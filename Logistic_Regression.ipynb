1. What is Simple Linear Regression (SLR)? Explain its purpose
- Simple Linear Regression is a statistical method used to model the relationship between one independent variable (X) and one dependent variable (Y).
- It fits a straight line that best describes how X influences Y.
- The purpose is to predict the value of Y based on X.
- It also helps understand the strength and direction of the relationship between variables.

2. What are the key assumptions of Simple Linear Regression?
- Linearity: The relationship between X and Y must be linear.
- Independence: Observations must be independent of each other.
- Homoscedasticity: The residuals must have constant variance.
- Normality of residuals: Residuals should follow a normal distribution.

4. Provide a real-world example where simple linear regression can be applied
- Predicting house price based on square footage.
- Predicting salary based on years of experience.
- Predicting sales based on advertising expenditure.
- Predicting temperature based on altitude.

5. What is the method of least squares in linear regression?
- It is a technique to estimate regression coefficients.
- It minimizes the sum of squared differences between actual and predicted Y values.
- Ensures the best-fitting line through the data.
- Produces unique values for slope and intercept.

6. What is Logistic Regression? How does it differ from Linear Regression?
- Logistic Regression predicts probabilities for classification problems (e.g., yes/no).
- It uses a sigmoid function to map predictions between 0 and 1.
- Linear Regression predicts continuous numeric values.
- Linear Regression outputs real numbers, while Logistic Regression outputs class probabilities.

7. Name and briefly describe three common evaluation metrics for regression models
- Mean Absolute Error (MAE) – Average of absolute differences between predicted and actual values.
- Mean Squared Error (MSE) – Average of squared prediction errors (penalizes large errors).
- R-squared (R²) – Measures how much variation in Y is explained by X.

8. What is the purpose of the R-squared metric in regression analysis?
- R² indicates the strength of the relationship between X and Y.
- Ranges from 0 to 1, where higher values show better fit.
- Shows the percentage of variability in Y explained by the model.
- Helps compare multiple regression models.

```python
# 9. Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept. (Include your Python code and output in the code box below.)

from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])

model = LinearRegression()
model.fit(X, y)

print("Slope (β1):", model.coef_[0])
print("Intercept (β0):", model.intercept_)
```

10. How do you interpret the coefficients in a simple linear regression model?
- Slope (β1): Amount Y changes for each 1-unit increase in X.
- Positive slope → Y increases with X; negative slope → Y decreases with X.
- Intercept (β0): Predicted value of Y when X = 0.
- Interpretation is meaningful only within the range of observed data.
